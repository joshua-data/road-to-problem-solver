# Contains Duplicate
* [Link](https://leetcode.com/problems/contains-duplicate/description/)

# My Approach

```python
from typing import List

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:

        # First, I created a list that only contains unique elements.
        nums_cleaned = list(set(nums))

        # Now, I compare the length of nums and nums_cleaned to return the answer.
            # [If there were duplicates in nums] The length would have changed during the process of extracting unique elements.
            # [If there were no duplicates in nums] The length would remain the same.
        return len(nums) != len(nums_cleaned)
```
