# Longest Consecutive Sequence
* [Link](https://leetcode.com/problems/longest-consecutive-sequence/description/)

# My Approach

```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        # If the list is empty, return 0!
        if nums is None:
            return 0

        # Convert to a set to avoid interference from duplicate elements!
        nums_set = set(nums)

        # Initialize the length of the longest consecutive sequence to 0!
        max_len = 0

        # Let's iterate through each element!
        for num in nums_set:

            # Only if it qualifies as the starting element of a sequence!
            if num - 1 not in nums_set:

                # Record the current number
                curr_num = num
                # Record the current sequence length
                curr_len = 1

                # Only while the next consecutive element exists!
                while curr_num + 1 in nums_set:
                    curr_num += 1  # Increment the current number by 1
                    curr_len += 1  # Increment the current sequence length by 1

                # Update the length of the longest consecutive sequence!
                max_len = max(max_len, curr_len)

        # Now return the answer!
        return max_len
```
