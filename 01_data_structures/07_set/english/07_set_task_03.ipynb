# Longest Substring Without Repeating Characters
* [Link](https://leetcode.com/problems/longest-substring-without-repeating-characters/description/)

# My Approach

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:

        # Variable explanation guide
        # === left_idx: Minimum index (pointer) of the substring with no duplicates
        # === right_idx: Maximum index (pointer) of the substring with no duplicates
        # === unique_saved_letters: Storage for letters to determine if left_idx should be updated
        # === max_len: The maximum length of a valid substring to return as the answer

        unique_saved_letters = set()
        left_idx = 0
        max_len = 0

        # Start by moving the right pointer one by one!
        for right_idx in range(len(s)):

            # If the right pointer element already exists in the letter storage?
            while s[right_idx] in unique_saved_letters:
                unique_saved_letters.remove(s[left_idx]) # Remove that element,
                left_idx += 1 # and move the left pointer to the right!

            # Add the right pointer element!
            unique_saved_letters.add(s[right_idx])
            # Now, the distance between the left and right pointers is the max length of a valid substring!
            max_len = max(max_len, right_idx - left_idx + 1)

        # Return the answer!
        return max_len
```
