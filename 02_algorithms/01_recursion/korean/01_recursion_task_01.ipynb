# Permutations
* [Link](https://leetcode.com/problems/permutations/description/)

# My Approach

```python
class Solution:
    def permute(self, nums):
        
        # 정답을 담을 순열 리스트
        permutations = []

        # 재귀 함수 정의
            # listed: 순열
            # unlisted: 후보
        def recursion(listed, unlisted):

            # 후보가 남아있을 경우
            if unlisted:
                for i in range(len(unlisted)):
                    recursion(
                        listed + [unlisted[i]], # 후보 하나씩 추가
                        unlisted[:i] + unlisted[i+1:] # 추가한 숫자를 제외한 나머지 후보들
                    )
            # 후보가 더 이상 없을 경우
            else:
                permutations.append(listed) # 순열이 완성되면 정답 목록에 추가
                return # 종료

        # Recursive Function 호출
        recursion([], nums)

        return permutations
```
