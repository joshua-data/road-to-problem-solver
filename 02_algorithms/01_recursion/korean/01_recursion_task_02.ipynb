# Combination Sum
* [Link](https://leetcode.com/problems/combination-sum/description/)

# My Approach

```python
class Solution:
    def combinationSum(self, candidates, target):

        # 정답을 담을 조합 리스트
        combinations = []
        # 각 조합
        combination = []

        # 재귀 함수 정의
        def rescursive(start_i, sum):

            # 아직 target에 도달하지 못한 경우            
            if sum < target:
                for i in range(start_i, len(candidates)):
                    num = candidates[i]
                    combination.append(num)
                    rescursive(i, sum + num)
                    combination.pop()

            # target에 도달한 경우
            elif sum == target:
                return combinations.append(combination[:])

            # target을 초과한 경우
            elif sum > target:
                return

        rescursive(0, 0)
        return combinations
```
